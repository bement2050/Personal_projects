```python
# on this script I am going to extract individual labs from google cloud quicklabs
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
list_of_labs_names=[]
list_of_labs_links=[]
```

```python
for i in list(range(23,180)):
    r=requests.get('https://www.qwiklabs.com/quests/{k}'.format(k=i))
    links=re.findall('<a data-type="Lab" href="(.*catalog)',r.text)
    names=re.findall('<a data-type="Lab" href=.*>(.*)</a>',r.text)
    for elements in links:
        list_of_labs_links.append(elements)
    for name in names:
        list_of_labs_names.append(name)
        
        
    
```

```python
list_of_full_link=["http://www.quiklabs.com"+a for a in list_of_labs_links]
```

```python
all_labs=pd.DataFrame(list_of_labs_names,list_of_full_link)
```
